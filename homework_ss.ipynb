### すべてのデータを取ってこれるようにfor文でまとめていく

```python

soup = BeautifulSoup(res.text,"html.parser")

data = []

# 物件名
properties = soup.find_all("div", attrs={"class":"property_unit"})

for property in properties:
    property_name = property.find("h2", attrs={"class":"property_unit-title"})
    property_name = property_name.text.replace("\n","")

    property_price = property.find("span", attrs={"class":"dottable-value"})
    property_price = property_price.text

    property_detailes = property.find_all("div", attrs={"class":"dottable-line"})

    property_distance = property_detailes[-4].find_all("dd")[1].text
    property_breadth = property_detailes[-3].find_all("dd")[1].text
    property_building_area = property_detailes[-2].find("dd").text

    details = {}

    details["物件名"] = property_name
    details["価格"] = property_price
    details["最寄りからの距離"] = property_distance
    details["間取り"] = property_breadth
    details["建物面積"] = property_building_area

    data.append(details)
```
